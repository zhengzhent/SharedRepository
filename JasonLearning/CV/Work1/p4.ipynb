```python
import numpy as np
import matplotlib.pyplot as plt

# 随机生成二值图像A和B
A = np.random.randint(0, 2, (4, 4))  # 4x4大小的随机二值矩阵
B = np.random.randint(0, 2, (4, 4))

# 计算交集和并集
intersection = np.logical_and(A, B)
union = np.logical_or(A, B)

plt.figure(figsize=(10, 5))

# 原始图像A和B
plt.subplot(2, 3, 1)
plt.imshow(A, cmap='binary')
plt.title('Image A')

plt.subplot(2, 3, 2)
plt.imshow(B, cmap='binary')
plt.title('Image B')

# 交集和并集
plt.subplot(2, 3, 4)
plt.imshow(intersection, cmap='binary')
plt.title('Intersection')

plt.subplot(2, 3, 5)
plt.imshow(union, cmap='binary')
plt.title('Union')

# 显示
plt.tight_layout()
plt.show()
```
