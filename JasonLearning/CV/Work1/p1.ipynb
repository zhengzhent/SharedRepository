```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

catpath = 'image\cat.jpg'
dogpath = 'image\dog.jpg'

catimage =  cv2.imread(catpath)
dogimage = cv2.imread(dogpath)
```

### 显示图像

```python
def cv_show(name,img):
    cv2.imshow(name,img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
cv_show('dog',dogimage)
```

### 转化为灰度图

```python
gray_img = cv2.cvtColor(dogimage, cv2.COLOR_BGR2GRAY)
# cv2.imshow('photo1',dogimage)
cv2.imshow('gray_dog',gray_img)
cv2.waitKey(0)
```

### 以png格式保存灰度图

```python
cv2.imwrite('processedImg/graydog.png', gray_img)
```

### 用4x4的排版显示原始图像以及R、G、B三个通道的图像；

```python
b,g,r = cv2.split(dogimage)
imgmerge = cv2.merge((r,g,b))
plt.imshow(imgmerge)       
```

```python
RGB_Image=cv2.merge([b,g,r])
RGB_Image = cv2.cvtColor(RGB_Image, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(12,12))
#显示各通道信息
plt.subplot(221)
plt.imshow(RGB_Image,'gray')
plt.title('RGB_Image')
plt.subplot(222)
plt.imshow(r,'gray')
plt.title('R_Channel')
plt.subplot(223)
plt.imshow(g,'gray')
plt.title('G_Channel')
plt.subplot(224)
plt.imshow(b,'gray')
plt.title('B_Channel')   
```
